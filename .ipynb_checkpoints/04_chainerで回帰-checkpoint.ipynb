## 必要なモジュールの読み込み

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 入力変数と教師データに分割

```python
x = df.iloc[:, :-1].values.astype('f')
t = df.iloc[:, -1].values.astype('f')
```

```python
x.shape
```

```python
t = t.reshape(len(t), 1)
```

## データセットの準備

```python
dataset = list(zip(x, t))
```

## 訓練データと検証データの分割

```python
import chainer
import chainer.functions as F
import chainer.links as L
```

```python
n_train = int(len(dataset) * 0.7)
n_train
```

```python
train, test = chainer.datasets.split_dataset_random(dataset, n_train, seed=0)
```

```python
# len(train)
```

```python
# len(test)
```

## モデルの定義

### L.Classgierを使う場合

```python
# 13->5->1
class NN(chainer.Chain):
    # モデルの構造
    def __init__(self, n_mid_units=5, n_out=1):
        super().__init__()
        with self.init_scope():
            self.fc1 = L.Linear(None, n_mid_units)
            self.fc2 = L.Linear(None, n_out)
            
            #Batch Normalization
            self.bn = L.BatchNormalization(13) #追記

    
    # 順伝搬
    def __call__(self, x):
        h = self.bn(x)
        h = self.fc1(h)
        h = F.relu(h)
        h = self.fc2(h)
        return h
```

```python
np.random.seed(0)
nn = NN(300)
```

```python
# 損失関数を平均二乗誤差に
model = L.Classifier(nn, lossfun=F.mean_squared_error)
```

```python
model.compute_accuracy = False
```

## 学習に必要な準備

```python
optimizer = chainer.optimizers.Adam()
optimizer.setup(model)
```

```python
batchsize = 70
```

```python
train_iter = chainer.iterators.SerialIterator(train, batchsize)
test_iter = chainer.iterators.SerialIterator(test, batchsize, repeat=False, shuffle=False)
```

```python
from chainer import training
```

```python
updater = training.StandardUpdater(train_iter, optimizer, device=-1)
```

```python
from chainer.training import extensions
```

```python
# エポックの数(学習回数の設定)
epoch = 500
```

```python
#trainerの宣言
trainer = training.Trainer(updater, (epoch, 'epoch'), out='result/housing')

# 検証データで評価
trainer.extend(extensions.Evaluator(test_iter, model, device=-1))

trainer.extend(extensions.LogReport(trigger=(1, 'epoch')))

# 1エポックごと（trigger)にtrainデータに対するlossとtestデータに対するloss, 経過時間を出力
trainer.extend(extensions.PrintReport([
    'epoch',
    'main/loss', 'validation/main/loss',
    'elapsed_time'
]), trigger=(1, 'epoch'))
```

## 学習の実行

```python
trainer.run()
```
